```python
df2.corr()

# coorelation values between two rows and two clumns always between -1 and plus 1
# ifthe coorelation between two columns are close to plus one thatmeans they are highly coorelated/ directly proportional
# ex:- if there are two columns x and y 
# if the coorelation between x and y is 0.5 then if xincreases then y increases by 0.5
# ex 2:- if the coorelation between x andy is -0.7 it means if youincrease 10 by factor 10 then y decrease by 7
# inverely proportional with the factor minus

# in the below dataframe the relation between fare and survival is 0.26 means as the fair increases the survival increases by 26%
# the coorelation between pcclass and survival is -0.33 that means if pcclass increases then survival decreases
# tht means pc class is 3, 2, 1 -> 1st class, 2nd class, 3rd class(that means numerically it is high but it is lower class)
# therefore the pc class increases survival rate decreases
```

```python
sns.heatmap(df2.corr())
```

```python
sns.heatmap(df2.corr(), annot=True, cmap='coolwarm')

# here annot = True is used to get the values on the graph and cmap = 'coolwarm' is used to change the colours
# here if it closely related to plus 1 then it is closely related,if it is related to -1 then it is inversely related 
# and if it is closely related to 0 then it is not related to anyone

# any matrix like coorelated can be plotted we can plot other types of matrix also covered later
```
